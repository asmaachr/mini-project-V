```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('df_cleaned.csv')
df.head()
```

```python
import matplotlib.pyplot as plt



# Plot the histogram
plt.hist(df['q1_count'], bins=15, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Q1 tokeninzed')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt



# Plot the histogram
plt.hist(df['q2_count'], bins=15, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Q2 tokeninzed')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt



# Plot the histogram
plt.hist(df['intersection_q1_q2_token_count'], bins=15, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of commen Q1 and Q2 tokeninzed')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'df' with columns 'intersection_q1_q2_token_count' and 'is_duplicate'
# Extract the data for the box plot
x_values = df['is_duplicate']
y_values = df['intersection_q1_q2_token_count']

# Set the size of the figure
plt.figure(figsize=(4, 3))  # Adjust the width and height as needed

# Create the box plot
sns.boxplot(x=x_values, y=y_values)

# Set the labels for x and y axes
plt.xlabel('is_duplicate')
plt.ylabel('intersection_q1_q2_token_count')

# Set the title of the plot
plt.title('Box Plot')

# Display the plot
plt.show()






```
